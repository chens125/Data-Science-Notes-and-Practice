```python
from sklearn.datasets import load_diabetes
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("insurance.csv")
data.head()
```

## 1. Plot Scatter Plot

```python
# PLot data
plt.figure()
sns.scatterplot(data=data, x="age", y="charges")
plt.show()
plt.close()
```

## 2. Fit a Model and Make Predictions

```python
# Select and reshape data
x = data.iloc[:, 0].values
y = data.iloc[:, -1].values

y = y.reshape(-1, 1)
x = x.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression

# Create an empty linear regression model
age_model = LinearRegression()

# Create the model using fit(x,y)
age_model.fit(x, y)

# Use model to make a prediction
y_pred = age_model.predict(x)
```

```python
# Predict an unknown value
unk_x = [[18]]

import numpy as np

x_pred = np.append(x, unk_x).reshape(-1, 1)
y_pred = age_model.predict(x_pred)

print(
    f"For someone who is {unk_x} years old, the average insurance charges is around {age_model.predict(unk_x)} pounds."
)
```

## 3. Plot Best-fit Line

```python
# Plot model and data
plt.scatter(x, y, color="b")
plt.plot(x, age_model.predict(x), color="r")
plt.title("Age vs Insurance Charges")
plt.xlabel("Age")
plt.ylabel("Insurance Charges")
plt.show()
```
