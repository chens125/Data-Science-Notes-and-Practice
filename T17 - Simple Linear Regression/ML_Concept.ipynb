# Machine learning

### 1. Defination
- ML is a computational methdology focused on deriving insights from data
- enables automated systems to adapt ad improve their performance over time

### 2. ML vs AI 
- ML is a a type of AI.
- AI: any program that completes a task in a way that can be considered human-like
- ML: AI programs that slove the task by learning without pre-programming

### 3. Input and Output
- Input: whatever we want the algorithem to lear from
eg:  features, independent variables,  explanatory variables
- Output: the outcome we want the algorithm to be able to preduce
eg: labels, predictions, dependent variables, respond variables

### 4. Supervised learning
- Annotation: machine learning from pairs of inputs and labels

**Regression**: a prediction task where a pragram learns to estimate and predict a continous output values
- Accuracy of model: r2,MSE, RMSE, MAE, MAPE

**Classification**: a program is trained to categorise input data into predifined classes 
Evaluation metrics: 

- Gini index: measures the impurity or disorder in a set of categorical data.

- Confusion matrix: summarises the model's performance by displaying counts of true positives, true negatives, false positives and false negatives.

- Precision: prportion

- F1 score: the harmonic mean of precision and recall, assesses the model's performance by considering both metrics
